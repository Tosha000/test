```python
def poliomin(params):
    
    T = params[0]
    T1 = T[0]
    T2 = T[1]
    S = params[1]
    Q = params[2]

    def matrix(T1=T1,T2=T2):
        m= [[0] * T1 for i in range(T2)]
        return m
    
    def coord(m,T1=T1,T2=T2):
        c = []
        for i in range(T2):
            for i2 in range(T1):
                if m[i][i2] == 1:
                    c.append([i,i2])
        return c
    
    def disp(m):
        for i in range(len(m)):
            print(m[i])
                    
    def right(m):
        m1 = matrix()
        c = coord(m)
        for i in range(len(c)):
            m1[c[i][0]][c[i][1]+1]=1
        return m1
    
    def down(m):
        m1 = matrix()
        c = coord(m)
        for i in range(len(c)):
            m1[c[i][0]+1][c[i][1]]=1
        return m1
    
    def up(m):
        m1 = matrix()
        c = coord(m)
        for i in range(len(c)):
            m1[c[i][0]-1][c[i][1]]=1
        return m1
    
    def all_pos(m,T1=T1,T2=T2):
        mn = []
        mn.append(m)
        step = T1 - 1 - coord(m)[-1][1]
        if step % 2 == 1:
            for i in range(step):
                if coord(m)[0][0] == 0 and coord(m)[0][1] % 2 == 0:
                    pass
                    while T1 -1 > coord(m)[-1][0]:
                        m = down(m)
                        mn.append(m)
                m = right(m)
                mn.append(m)
                while T1 - coord(m)[0][0] < T1:
                    m = up(m)
                    mn.append(m)
        if step % 2 == 0:
            for i in range(step):
                if coord(m)[0][0] == 0 and coord(m)[0][1] % 2 == 0:
                    pass
                    while T1 -1 > coord(m)[-1][0]   :
                        m = down(m)
                        mn.append(m)
                m = right(m)
                mn.append(m)
                while T1 - coord(m)[0][0] < T1:
                    m = up(m)
                    mn.append(m)
                if coord(m)[0][0] == 0 and coord(m)[0][1] % 2 == 0:
                    pass
                    while T1 -1 > coord(m)[-1][0]   :
                        m = down(m)
                        mn.append(m)
        return mn
    
    def delete_double(m):
        m_new = []
        [m_new.append(i) for i in m if i not in m_new]
        return m_new
    
    def summ(m1,m2):
        result = [[m1[i][j] + m2[i][j]  for j in range(len(m1[0]))] for i in range(len(m1))] 
        return result
        
    def prov(mn1,mn2):
        mn=[]
        for i in range(len(mn1)):
            for j in range(len(mn2)):
                x = summ(mn1[i],mn2[j])
                s=0
                for k in range(len(x)):
                    for k1 in range(len(x[k])):
                        if x[k][k1] > 1:
                            s  = s + 1
                if s == 0:
                    mn.append(summ(mn1[i],mn2[j]))         
        return mn
    
    def pryam(S):
        m=matrix()
        for i in range(S[1]):
            for i2 in range(S[0]):
                m[i][i2] = m[i][i2] + 1
        return m
    def tpryam(S):
        m=matrix()
        for i in range(S[0]):
            for i2 in range(S[1]):
                m[i][i2] = m[i][i2] + 1
        return m 
    
    def final(S=S):
        all_S = []
        for i in range(len(S)):
            for i2 in range(S[i][1]):
                all_S.append(S[i][0])
                
        Sx=[]
        if len(S) == 1:
            Sx.append(all_pos(pryam(S[0][0])) + all_pos(tpryam(S[0][0])))
            
        if len(S) == 2:
            Sx.append(prov(all_pos(pryam(S[0][0])) + all_pos(tpryam(S[0][0])),all_pos(pryam(S[1][0])) + all_pos(tpryam(S[1][0]))))
            
        if len(S) > 1:
            pos1 = all_pos(pryam(S[0][0])) + all_pos(tpryam(S[0][0]))
            pos2 = all_pos(pryam(S[1][0])) + all_pos(tpryam(S[1][0]))
            Sn = prov(pos1,pos2)
        
            if len(S) > 2:
                for i in range(len(S)):
                    if i > 1:
                        x = all_pos(pryam(S[i][0])) + all_pos(tpryam(S[i][0]))
                        for i1 in range(len(Sn)):
                            for i2 in range(len(x)):
                                if len(prov(Sn,[x[i2]])) > 0:
                                    for k in range(len(prov(Sn,[x[i2]]))):
                                        Sx.append(prov(Sn,[x[i2]])[k])
                                        
        Sx = delete_double(Sx)
        return all_S,Sx
                                        
    print(final())
    
#переделать all_pos, пробелма с пропуском итерации вниз

 
                
poliomin([(4,4),[[(1,4),2],[(1,2),1]],[(3,3),1]])
```
